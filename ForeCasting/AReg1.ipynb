```python
import pandas as pd
import numpy as np
from matplotlib import pyplot
from statsmodels.tsa.ar_model import AutoReg
```

```python
#set date column as index and telling pandas to treat date as index and not string 
df=pd.read_csv('MaunaLoaDailyTemps.csv',index_col='DATE',parse_dates=True) 
#drop missing value
df=df.dropna() 
#number of rows and columns
print('Shape of data',df.shape) 
df.head(4)
```

```python
dataset=df['AvgTemp']
print(dataset.shape)
dataset.head()
dataset.plot(figsize=(12,5))
```

```python
#checing for stationarity
from statsmodels.tsa.stattools import adfuller
dftest = adfuller(dataset, autolag = 'AIC')

print("1. ADF : ",dftest[0])
print("2. P-Value : ", dftest[1])
print("3. Num Of Lags : ", dftest[2])
print("4. Num Of Observations Used For ADF Regression and Critical Values Calculation :", dftest[3])
print("5. Critical Values :")
for key, val in dftest[4].items():
    print("\t",key, ": ", val)

if dftest[1] <0.05:
    print('as the datasets p value:',dftest[1],' is less than 0.05, the dataset is stationary')
#if p value <0.05, then data is stationary
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
pacf=plot_pacf(dataset,lags=25)
act=plot_acf(dataset,lags=25)
```

```python
#Split Data into Training and Testing
print(dataset.shape)
train=dataset.iloc[:-30]  #rest other data rows for training
test=dataset.iloc[-30:] #keeping last 30 rows value for testing
print(train.shape,test.shape)
```

```python
import warnings
warnings.filterwarnings("ignore")

model=AutoReg(train,lags=2).fit() 

print(model.summary())
```

```python
#testing

start=len(train)
end=len(train)+len(test)-1
pred=model.predict(start=start,end=end,dynamic=False).rename('ARtesting')
#if indexing is not in terms of date and we want in it terms of date, then we have explicity tell it
index_future_dates=pd.date_range(start='2018-12-01',end='2018-12-30')
pred.index=index_future_dates
print(pred)
pred.plot(figsize=(12,5),legend=True)   
test.plot(figsize=(12,5),legend=True)
```

```python
test.mean()
```

```python

from sklearn.metrics import mean_squared_error
from math import sqrt
rmse=sqrt(mean_squared_error(pred,test))
print(rmse)
```
